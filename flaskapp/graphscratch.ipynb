```python
fig, ax = plt.subplots(figsize=(10, 10))

seed_order = [n for n in node_ordering if n in ag.nearby.seed_nodes]
targ_order = [n for n in node_ordering if n in ag.target.seed_nodes]

r_near = 10
r_targ = 1

initial_pos = {
    key: (
        r_near * np.cos(idx * np.pi / len(seed_order)),
        r_near * np.sin(idx * np.pi / len(seed_order))
    )
    for idx, key in enumerate(seed_order)
}

initial_pos.update({
    key: (
        r_targ * np.cos(idx * np.pi / len(targ_order)),
        -r_targ * np.sin(idx * np.pi / len(targ_order))
    )
    for idx, key in enumerate(targ_order)
})


pos = nx.spring_layout(
    G,
    k=0.9,
    pos=initial_pos,
    fixed=initial_pos.keys()
)

pos_near = {
    node: pos
    for node, pos in pos.items()
    if node in ag.nearby.seed_nodes
}

pos_targ = {
    node: pos
    for node, pos in pos.items()
    if node in ag.target.seed_nodes
}

pos_inter = {
    node: pos
    for node, pos in pos.items()
    if node not in pos_near and node not in pos_targ
}

nx.draw_networkx_nodes(G.subgraph(pos_near), pos_near, node_color="orange", node_size = 200)
nx.draw_networkx_nodes(G.subgraph(pos_targ), pos_targ, node_color="purple", node_size = 200)
nx.draw_networkx_nodes(G.subgraph(pos_inter), pos_inter, node_color="grey", node_size = 50)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=G.edges, arrows=True, alpha=0.2)
plt.show()
```
